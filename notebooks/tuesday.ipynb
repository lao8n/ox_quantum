```python
from qiskit import QuantumCircuit
# Create circuit with n qubits
circ = QuantumCircuit(3)
# State preparation
                    # q0|0> 
circ.ry(pi/4, 0)    # q0|+>
# circ.rx(-pi/2, 0) # q0|R>
                    # q1|0>
                    # q2|0>
# Apply gates to qubits
circ.h(1)
# circ.u(theta, phi, lam, 0)
circ.cx(1, 2)       # q1 control, q2 target
circ.cx(0, 1)       # q1 control, 10 target
circ.h(0)
circ.cx(1, 2)       # q1 control, q2 target
circ.cz(0, 2)       

# Visualize
display(circ.draw("mpl", initial_state=True))
plot_bloch_multivector(circ)

# Copies the state from q0 to q2
```
